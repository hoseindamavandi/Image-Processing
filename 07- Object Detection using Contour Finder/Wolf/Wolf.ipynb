```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('wolf.jpg', 0)
plt.imshow(img, cmap="gray")
```

```python
_ ,img_threshold = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)
plt.imshow(img_threshold, cmap="gray")
```

```python
contours, _ = cv2.findContours(img_threshold, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

for c in contours:
    if len(c) > 1000:
        x, y, w, h = cv2.boundingRect(c)
        cv2.rectangle(img, (x,y), (x+w , y+h), (127,127,127), 8)
        
plt.imshow(img, cmap="gray")
```
